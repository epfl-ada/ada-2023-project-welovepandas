```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
users = pd.read_csv("../dataset/BeerAdvocate/users.csv")

users["is_american"] = users["location"].str.contains("United States")

display(users.head())
```

```python
plt.figure(figsize=(30, 8))

sns.histplot(
    data=users,
    x="location",
)

plt.xticks(rotation=90)
```
